```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Load data from 'exercise.mat'
data = loadmat('Library/exercise.mat')

X = data['X']
Y = data['Y']
map_image = data['map']
obstacleMap = data['obstacleMap']
xLM = data['xLM']
yLM = data['yLM']
Xvec = data['Xvec']
Yvec = data['Yvec']

# Resolution of the map
mapRes = 10  # meters

# Example coordinates
x0 = 0  # meters
y0 = 0  # meters

# Compute indices
j = int((x0 - X[0, 0] + mapRes) / mapRes) - 1   # column index
i = int((Y[0, 0] - y0 + mapRes) / mapRes) - 1   # row index

# Check values
print(x0 == X[i, j], y0 == Y[i, j])

# Squeeze Xvec and Yvec
Xvec = np.squeeze(Xvec)     # both these variables must be one-dimensional vectors
Yvec = np.squeeze(Yvec)

# Plotting the map of the environment in grayscale
plt.figure()
plt.imshow(map_image, cmap='gray', extent=(Xvec[0], Xvec[-1], Yvec[-1], Yvec[0]))
plt.axis('on')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Obstacle map in grayscale
plt.figure()
plt.imshow(obstacleMap, cmap='gray', extent=(Xvec[0], Xvec[-1], Yvec[-1], Yvec[0]))
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Landmarks on grayscale map
plt.figure()
plt.imshow(map_image, cmap='gray', extent=(Xvec[0], Xvec[-1], Yvec[-1], Yvec[0]))
plt.plot(xLM, yLM, 'yo', markersize=5, linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
